#**Statistics Advance Part 1 Assignment(Questions and Answers):**



1.**What is a random variable in probability theory?**

Ans: A random variable is a numerical outcome of a random phenomenon.

2.**What are the types of random variables?**

Ans: There are two types: discrete and continuous random variables.

3.**What is the difference between discrete and continuous distributions?**

Ans: Discrete distributions have countable outcomes; continuous distributions have infinite outcomes within a range.

4.**What are probability distribution functions (PDF)?**

Ans: PDF describes the likelihood of a random variable to take on a particular value.

5.**How do cumulative distribution functions (CDF) differ from probability distribution functions(PDF)?**

Ans: CDF gives the probability that a variable is less than or equal to a value; PDF gives the probability at a specific value.

6.**What is a discrete uniform distribution?**

Ans: A discrete uniform distribution has equal probability for all outcomes in a finite set.

7.**What are the key properties of a Bernoulli distribution?**

Ans: It has only two outcomes (0 or 1) with a single parameter p representing the probability of success.

8.**What is the binomial distribution, and how is it used in probability?**

Ans: It models the number of successes in a fixed number of trials, each with the same probability of success.

9.**What is the Poisson distribution and where is it applied?**

Ans: It models the number of events in a fixed interval of time/space with a constant mean rate.

10.**What is a continuous uniform distribution?**

Ans: It assigns equal probabilities over a continuous range.

11.**What are the characteristics of a normal distribution?**

Ans: Symmetrical, bell-shaped, defined by mean and standard deviation.

12.**What is the standard normal distribution, and why is it important?**

Ans: A normal distribution with a mean of 0 and a standard deviation of 1; used to standardize data.

13.**What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Ans: CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases.

14.**How does the Central Limit Theorem relate to the normal distribution?**

Ans: It justifies the use of normal distribution in many real-world problems, even when the population isn't normal.

15.**What is the application of Z statistics in hypothesis testing?**

Ans: Z statistics help test hypotheses by comparing sample and population means.

16.**How do you calculate a Z-score, and what does it represent?**

Ans: Z = (X - mu) / sigma, it shows how many standard deviations a value is from the mean.

17.**What are point estimates and interval estimates in statistics?**

Ans: Point estimate gives a single value; interval estimate gives a range with a confidence level.

18.**What is the significance of confidence intervals in statistical analysis?**

Ans: Confidence intervals give a range in which the population parameter is expected to lie.

19.**What is the relationship between a Z-score and a confidence interval?**

Ans: Z-score determines how far a value is from the mean, aiding in confidence interval calculation.

20.**How are Z-scores used to compare different distributions?**

Ans: Z-scores standardize different datasets for comparison.

21.**What are the assumptions for applying the Central Limit Theorem?**

Ans: Random sampling, independent observations, sample size large enough (n > 30).

22.**What is the concept of expected value in a probability distribution?**

Ans: Expected value is the long-term average or mean of a random variable.

23.**How does a probability distribution relate to the expected outcome of a random variable?**

Ans: It provides the average outcome expected over many repetitions of the experiment.
